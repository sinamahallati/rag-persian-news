# ۱ – آماده‌سازی داده (Persian News Dataset)

در این نوت‌بوک، دیتاست خبری فارسی را از پوشه‌ی `data/raw` می‌خوانید، یک زیرمجموعه‌ی کوچک و قابل‌مدیریت از آن می‌سازید، داده‌ها را کمی تمیز می‌کنید و نتیجه را در مسیر `data/processed/news_subset.csv` ذخیره می‌کنید.

```python
import pandas as pd
from pathlib import Path
```

```python
raw_path = Path('/Users/sinamahallati/Desktop/GenAI/RAG/data/raw/archive_v5.csv')
subset_path = Path('/Users/sinamahallati/Desktop/GenAI/RAG/data/raw/news_subset.csv')
```

```python
df = pd.read_csv(raw_path)
df.head()    
```

```python
print(df.shape)
print(df.isnull().sum())

min_chars = 50

df["body"] = df["body"].fillna("").astype(str).str.strip()
df = df[df["body"].str.len() >= min_chars].copy()

df = df.dropna().copy()

print(df.shape)
print(df.isnull().sum())
```

```python
N = 20000
df_subset = df.sample(n=min(N, len(df)), random_state=42).copy()

print(df_subset.shape)
df_subset.head()
```

```python
def fa_normalize(text: str) -> str:
    if pd.isna(text):
        return ""
    text = str(text)
    text = text.replace("ي", "ی").replace("ك", "ک")
    return text

df_subset["body"] = df_subset["body"].apply(fa_normalize)
```

```python
df_subset.to_csv(subset_path, index=False)
print("Saved to:", subset_path)
```
